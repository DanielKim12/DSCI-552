### (c) Simple Linear Regression

```python
# Function to plot the simple linear regression graph
def plot_graph(x, y, y_pred, outlier_pt, col_name, response_name="PE"): #response vaiable is PE
    fig, axes = plt.subplots(figsize=(10, 6))
    plt.scatter(x, y, color='green', alpha=0.2, label='Data Points') #setting alpha to 0.2 to make the points transparent
    plt.scatter(x[outlier_pt], y[outlier_pt], color='red', label='Outliers') #plotting outliers 
    plt.plot(x, y_pred, color='black', label='Regression') 
    plt.title(f'{response_name} vs {col_name}') #for comparison of the response variable and the predictor variable
    plt.xlabel(col_name)
    plt.ylabel(response_name)
    plt.legend()
    plt.show()

# write summary and find the outliers 
def report_linear_outliers(col_name):
    # refresh the results and outlier data as empty lists
    results = []
    outlier_data = []
    linear_coef = []
    # Get the predictor and response variables
    x = df[col_name].to_numpy().reshape(-1, 1)  #convert it one-dimensional array
    y = df['PE'].to_numpy()  # Response variable 'PE'
    # Fit the simple regression model
    model = LinearRegression()
    model.fit(x, y)
    # Get the coefficient and intercept
    coef = model.coef_[0]  # Linear coefficient, beta_0
    intercept = model.intercept_ #slopes beta_1, beta_2, ...
    # Predict the response variable
    y_pred = model.predict(x)
    # Calculate mse and r2
    mse = mean_squared_error(y, y_pred)
    r2 = r2_score(y, y_pred)

    # Append results to the list
    results = [
        ["Intercept", intercept],
        ["Coefficient", coef],
        ["Mean Squared Error", mse],
        ["R-squared", r2]
    ]
    print(f"\nRegression Summary for {col_name}:\n")
    print(tabulate(results, headers=["Metric", "Value"], tablefmt="grid"))
    
    # cooks distance
    x_const = sm.add_constant(x)  # Add constant for intercept 
    model_sm = sm.OLS(y, x_const).fit()  # Fit the OLS model using statsmodels
    influence = model_sm.get_influence()  # Get influence measures 
    cooks_d = influence.cooks_distance[0]  # Cook's distance values
    # Use the standard threshold for Cook's distance: 4/n
    outlier_pt = np.where(cooks_d > 4 / len(x))[0]  # Detect outliers, if Cook's distance > 4/n
    # Iterate through outlier indices and append to list
    for idx in outlier_pt:
        outlier_data.append([idx, df[col_name].iloc[idx], df['PE'].iloc[idx]]) 
    # Convert it to a DataFrame
    outlier_df = pd.DataFrame(outlier_data, columns=['Index', col_name, 'PE'])
    print(f"\nOutliers for {col_name}:\n")
    # print(tabulate(outlier_df, headers='keys', tablefmt='grid'))
    # Use a simple HTML table with a scrollable output
    html_outliers = outlier_df.to_html()
    display(HTML(f"<div style='max-height: 300px; overflow-y: scroll;'>{html_outliers}</div>"))
    plot_graph(x.flatten(), y, y_pred, outlier_pt, col_name) #plot outliers 
    return outlier_pt, x.flatten(), y
```

### (d) Multiple Regression

```python
# import statsmodels.formula.api as smf

# https://www.datarobot.com/blog/multiple-regression-using-statsmodels/
# multiple predictors
res = smf.ols('PE ~ V + AP + AT + RH', data=df).fit()
print(res.summary())
```

We can reject the null hypothesis for all variables (V, AP, AT, RH) since their p-values are 0, indicating that all of these predictors are statistically significant. Additionally, the model's R-squared value of 0.929 demonstrates a strong fit, this tells that this model is better predictor for the PE. 

### (e) 1c Compare to 1d

```python
predictors = ['AT', 'V', 'AP', 'RH']
univariate_coefs = []
for col in predictors:
    x_simple = df[col].to_numpy().reshape(-1, 1)
    x_simple_const = sm.add_constant(x_simple)  # Add constant for intercept
    model_simple = sm.OLS(df['PE'], x_simple_const).fit()
    univariate_coefs.append(model_simple.params[1])  # Store the coefficient of the predictor

# Fit the multiple regression model using statsmodels
X_multiple = df[predictors]
X_multiple_const = sm.add_constant(X_multiple)  # Add constant for intercept
model_multiple = sm.OLS(df['PE'], X_multiple_const).fit()

# multiple regression coefficients
multiple_coefs = model_multiple.params[1:]  # Exclude the intercept

# Plot the comparison with custom legend
fig, axes = plt.subplots(figsize=(10, 6))
# https://www.geeksforgeeks.org/style-plots-using-matplotlib/
with plt.style.context('bmh'):
    # displaying two decimal places for the coefficients as legend 
    # https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.axes.Axes.scatter.html
    axes.scatter(univariate_coefs[0], multiple_coefs[0], label=f"AT [{univariate_coefs[0]:.2f}, {multiple_coefs[0]:.2f}]", color='blue')
    axes.scatter(univariate_coefs[1], multiple_coefs[1], label=f"V [{univariate_coefs[1]:.2f}, {multiple_coefs[1]:.2f}]", color='red') 
    axes.scatter(univariate_coefs[2], multiple_coefs[2], label=f"AP [{univariate_coefs[2]:.2f}, {multiple_coefs[2]:.2f}]", color = 'black')
    axes.scatter(univariate_coefs[3], multiple_coefs[3], label=f"RH [{univariate_coefs[3]:.2f}, {multiple_coefs[3]:.2f}]", color = 'yellow')

    # Annotate the points with the predictor names
    for i, predictor in enumerate(predictors):
        axes.annotate(predictor, (univariate_coefs[i], multiple_coefs[i]), textcoords="offset points", xytext=(10,2), ha='center')

    axes.set_title(f"Univariate vs Multiple Regression")
    axes.set_xlabel('Univariate Coefficients')
    axes.set_ylabel('Multiple Coefficients')
    # Add grid lines and a legend
    axes.legend()
    axes.axhline(0, color='black', linewidth=1)
    axes.axvline(0, color='black', linewidth=1)
    axes.grid(True)

# Show the plot
plt.show()
```

### (f) Nonlinear Association

```python
def report_nonlinear(col_name):
    # https://stackoverflow.com/questions/16665833/statsmodels-specifying-non-linear-regression-models-using-patsy 
    # usage of I() operator to specify the polynomial terms
    eq = f'PE ~ {col_name} + I({col_name}**2) + I({col_name}**3)'
    res = smf.ols(formula=eq, data=df).fit()
    print(res.summary())
```

```python
report_nonlinear('AT')
```

We can reject the null hypothesis as p-value (0.000) is less than alpha (5%), thus AT(predictor) shows non-linear association with PE (response variable).

```python
report_nonlinear('V')
```

We fail to reject the null hypothesis as p-value is greater than alpha, thus V(predictor) shows no significant non-linear association with PE (response variable).

```python
report_nonlinear('AP')
```

We can reject the null hypothesis as p-value (0.000) is less than alpha (5%), thus AP(predictor) shows non-linear association with PE (response variable).

```python
report_nonlinear('RH')
```

We can reject the null hypothesis as p-value (0.000) is less than alpha (5%), thus RH(predictor) shows non-linear association with PE (response variable).

### (g) Interactions of Predictors

```python
# Full model with all pairwise interaction terms between AP, V, AT, and RH, cited below
eq = 'PE ~ AT + V + AP + RH + AT:V + AT:AP + AT:RH + V:AP + V:RH + AP:RH'
# Fit the model using the formula
res = smf.ols(formula=eq, data=df).fit()
# Print the summary of the model
print(res.summary())
```

V, AP, and RH predictors are statistically significant as their p-values are less than alpha. Additionally, the pairwise interaction terms AT:V, AT:RH, V:AP, and AP:RH are statistically significant as their p-values are also less than alpha.

### (h) Improvement

```python
# Split the data into training and testing sets, train data is randomly selected 70% of the data
x = df[['AT', 'V', 'AP', 'RH']] #predictor variables
y = df['PE'] #response variable
#https://stackoverflow.com/questions/28064634/random-state-pseudo-random-number-in-scikit-learn
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42) #randomly selected 70% of the data, state is set to 42
analysis = [] #to store the results of the analysis
```

```python
results = []
eq = 'PE ~ AT + V + AP + RH'
# Fit the model using the formula
res = smf.ols(formula=eq, data=df).fit()

# Predict the response variable using the model, split between test and train data
y_pred_test = res.predict(x_test)
y_pred_train = res.predict(x_train)
# Calculate the mean squared error for the test and train data
mse_test = mean_squared_error(y_test, y_pred_test)
mse_train = mean_squared_error(y_train, y_pred_train)

results.append(["Min_Test_MSE", mse_test])
results.append(["Min_Train_MSE", mse_train])
#print results using tabulate
print(tabulate(results, headers=["Metric", "Value"], tablefmt="grid"))
# for analysis 
analysis.append ({
    'Model': 'Simple Linear Regression with all predictors',
    'Min_Test_MSE': mse_test,
    'Min_Train_MSE': mse_train
})
```

```python
results = []
eq = 'PE ~ AT + V + AP + RH + AT:V + AT:AP + AT:RH + V:AP + V:RH + AP:RH + np.power(AT, 2) + np.power(V, 2) + np.power(AP, 2) + np.power(RH, 2)'  # Interactions and Quadratic terms
# Fit the model using the formula
res = smf.ols(formula=eq, data=df).fit()

# Predict the response variable using the model, split between test and train data
y_pred_test = res.predict(x_test)
y_pred_train = res.predict(x_train)

results.append(["Min_Test_MSE", mean_squared_error(y_test, y_pred_test)])
results.append(["Min_Train_MSE", mean_squared_error(y_train, y_pred_train)])
#print results using tabulate
print(tabulate(results, headers=["Metric", "Value"], tablefmt="grid"))
analysis.append ({
    'Model': 'Multiple including Interactions and quadratic terms Linear Regression',
    'Min_Test_MSE': mean_squared_error(y_test, y_pred_test),
    'Min_Train_MSE': mean_squared_error(y_train, y_pred_train)
})
```

### (i) KNN

```python
#copied from hw 1 for knn 
# import numpy as np
# from sklearn.neighbors import KNeighborsRegressor

test_err = []
train_err = []
mse_test_data = []
mse_train_data = []
results = []
k_val = range(1,101,1)
best_k = 0

for k in k_val: 
    # using kneighbors regressor to fit the model
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(x_train, y_train) #train x and y 
    #make predictions on train and test data 
    y_pred_train = knn.predict(x_train)
    y_pred_test = knn.predict(x_test) 
    mse_test = mean_squared_error(y_test, y_pred_test)
    mse_train = mean_squared_error(y_train, y_pred_train)
    
    #calculating the train errors 
    train_error = 1 - knn.score(x_train, y_train)
    train_err.append(train_error)
    mse_test_data.append(mse_test)
    
    #calculating the train errors 
    test_error = 1 - knn.score(x_test, y_test)
    test_err.append(test_error)
    mse_train_data.append(mse_train)

#get min test mse along with best k value
min_mse_test = min(mse_test_data)
#https://www.geeksforgeeks.org/numpy-argmin-python/  for np.argmin
best_k = k_val[mse_test_data.index(min_mse_test)]
min_test_idx = np.argmin(mse_test_data)
min_mse_train = mse_train_data[min_test_idx]

#want to append the results to the table best k and min mse
print("Without Normalization")
results.append(["Best K", best_k])
results.append(["Min_Test_MSE", min_mse_test])
results.append(["Min_Train_MSE", min_mse_train])
#print results using tabulate
print(tabulate(results, headers=["Metric", "Value"], tablefmt="grid"))

# Plot the training and testing errors in terms of 1/k, as k increases the error decreases
inverse_k = 1/np.array(k_val)
plt.figure(figsize=(10,6))
plt.plot(inverse_k, train_err, label='Without Normalized Training Error')
plt.plot(inverse_k, test_err, label='Without Normalized Testing Error')
plt.xlabel('1/K values')
plt.ylabel('Error rates')
plt.title('Without Normalization Training and Testing Error vs 1/K') 
plt.legend()
plt.show()

analysis.append ({
    'Model': 'KNN Regression without Normalization',
    'Min_Test_MSE': min_mse_test,
    'Min_Train_MSE': min_mse_train
})
```

```python
test_err = []
train_err = []
mse_test_data = []
mse_train_data = []
results = []
k_val = range(1,101,1)
best_k = 0

# normalize the data
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

for k in k_val: 
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(x_train_scaled, y_train) #train x and y 
    #make predictions on train and test data 
    y_pred_train = knn.predict(x_train_scaled)
    y_pred_test = knn.predict(x_test_scaled) 

    mse_test = mean_squared_error(y_test, y_pred_test)
    mse_train = mean_squared_error(y_train, y_pred_train)
    
    #calculating the train errors 
    train_error = 1 - knn.score(x_train_scaled, y_train)
    train_err.append(train_error)
    mse_test_data.append(mse_test)
    
    #calculating the train errors 
    test_error = 1 - knn.score(x_test_scaled, y_test)
    test_err.append(test_error)
    mse_train_data.append(mse_train)

#get min test mse along with best k value
min_mse_test = min(mse_test_data)
#https://www.geeksforgeeks.org/numpy-argmin-python/ 
best_k = k_val[mse_test_data.index(min_mse_test)]
min_test_idx = np.argmin(mse_test_data)
min_mse_train = mse_train_data[min_test_idx]

#want to append the results to the table best k and min mse
print("With Normalization")
results.append(["Best K", best_k])
results.append(["Min_Test_MS", min_mse_test])
results.append(["Min_Train_MSE", min_mse_train])
#print results using tabulate
print(tabulate(results, headers=["Metric", "Value"], tablefmt="grid"))

inverse_k = 1/np.array(k_val)
plt.figure(figsize=(10,6))
plt.plot(inverse_k, train_err, label='Without Normalized Training Error')
plt.plot(inverse_k, test_err, label='Without Normalized Testing Error')
plt.xlabel('1/K values')
plt.ylabel('Error rates')
plt.title('With Normalization Training and Testing Error vs 1/K')
plt.legend()
plt.show()

analysis.append ({  
    'Model': 'KNN Regression with Normalization',
    'Min_Test_MSE': min_mse_test,
    'Min_Train_MSE': min_mse_train
})
```

## 2. ISLR: 2.4.1

#### For each of parts (a) through (d), indicate whether we would generally expect the performance of a flexible statistical learning method to be better or worse than an inflexible method. Justify your answer.

### (a) The sample size n is extremely large, and the number of predictors p is small.

Flexible method will perform better as it allows advantage of a large sample size n, where large sample size greatly reduce the risk of overfitting. 

### (b) The number of predictors p is extremely large, and the number of observations n is small.

A flexible method will perform worse, as mentioned above. While a large sample size is beneficial for flexible methods, they tend to overfit when the number of observations is small.

### (c) The relationship between the predictors and response is highly non-linear.

Flexible method will perform better with more degrees of freedom as the relationship between the predictors and response is highly non-linear.

### (d) The variance of the error terms, i.e. $σ^2$ = Var(ε), is extremely high.


Flexible method will perform worse since it will try to fit the unexplained error terms (noise in the error terms and increase variance).

## 3. ISLR: 2.4.7

### (a) Compute the Euclidean distance between each observation and the test point, X1 = X2 = X3 = 0.

```python
ans = []
data = [[0,3,0], [2,0,0], [0,1,3], [0,1,2], [-1,0,1], [1,1,1]] #data points x1, x2, x3
index = [1,2,3,4,5,6] #observation number
dummy = [[0, 0, 0]] #comparing the data points for nearest neighbor
Y = ["Red", "Red", "Red", "Green", "Green", "Red"]
# https://stackoverflow.com/questions/72759308/calculating-euclidean-distance-in-python
dist = pairwise.euclidean_distances(data, dummy) #calculating the distance between the data points and the dummy point
# get dist of all points to the dummy point
for i in range(len(data)):
    tmp = dist[i][0]
    #append to the ans two decimal places include Y as well 
    ans.append([index[i], data[i], round(tmp, 2), Y[i]])
#print results using tabulate
print(tabulate(ans, headers=["Obs.", "X1, X2, X3", "Euclidean Distance", "Y"], tablefmt="grid"))
```

### (b) What is our prediction with K = 1? Why?

From the table, we can see that the closest neighbor (K = 1), based on the Euclidean distance, is Observation 5, which is located at [-1, 0, 1] with a distance of 1.41. Since its class label is Green, our prediction is Green.

### (c) What is our prediction with K = 3? Why?

From the table, we can predict that for K = 3, the class will be Red. Based on the Euclidean distance, Observation 2, which is located at [2, 0, 0] with a distance of 2, will be the third nearest neighbor. The sequence is as follows: [-1, 0, 1] -> [1, 1, 1] -> [2, 0, 0]. Therefore, our prediction is Red.

### (d) If the Bayes decision boundary in this problem is highly non-linear, then would we expect the best value for K to be large or small? Why?

If the Bayes decision boundary in this problem is highly non-linear, then the decision boundary will indicate a flexible method, as it performs better. Thus, the best value for K will be small, as it allows for a flexible and adaptive boundary to closely fit the non-linear structure of the data.

##### Prompt: how to perform simple linear regression? 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Example data
data = {'X': [1, 2, 3, 4, 5], 'Y': [2, 4, 5, 4, 5]}
df = pd.DataFrame(data)

# Split the data into predictor (X) and response (Y)
X = df[['X']]  # Predictor (2D)
Y = df['Y']  # Response (1D)

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, Y)

# Make predictions
Y_pred = model.predict(X)

# Evaluate the model
mse = mean_squared_error(Y, Y_pred)
r2 = r2_score(Y, Y_pred)
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")

# Coefficients
print(f"Intercept (b0): {model.intercept_}")
print(f"Slope (b1): {model.coef_[0]}")

# Visualize the data and regression line
plt.scatter(X, Y, color='blue', label='Data points')
plt.plot(X, Y_pred, color='red', label='Regression line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```

##### prompt: what does it mean to run full linear regression moodel with all pariwise interaction terms? like AP and V? 

response: If you have predictors AP, V, AT, and RH, the formula for a full model with all pairwise interaction terms looks like this:
PE ~ AP * V * AT * RH
The * operator automatically expands to include all linear terms and pairwise interaction terms. So the formula expands to:
PE ~ AP + V + AT + RH + AP:V + AP:AT + AP:RH + V:AT + V:RH + AT:RH

```python
#example code using statsmodels
import statsmodels.formula.api as smf
# Full model with all pairwise interaction terms between AP, V, AT, and RH
formula = 'PE ~ AP + V + AT + RH + AP:V + AP:AT + AP:RH + V:AT + V:RH + AT:RH'
# Fit the model using the formula
model = smf.ols(formula=formula, data=df).fit()
# Print the summary of the model
print(model.summary())
```

response: Include interaction terms: For every pair of predictors, you model how their combined effect influences the response variable (in this case, PE). This means you're not just considering the individual effects of AT, V, AP, and RH on PE, but also how the effect of one predictor depends on the value of another predictor. Include quadratic terms: These terms capture non-linear relationships by adding the squared value of each predictor. Quadratic terms allow for a curved relationship between the predictor and the response. For example, the effect of AT on PE may not be linear, but instead, PE might increase with AT up to a certain point, and then decrease as AT increases further.

PE ~ AT + V + AP + RH  # Linear terms

  + V:AP + V:RH  # Interaction terms for V
  + AP:RH  # Interaction term for AP and RH
